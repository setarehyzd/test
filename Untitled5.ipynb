```python
import re

# Define regex patterns with proper capturing groups
# amount_pattern = r"(?:Total Amount Due|Amount Due|Balance Due|Grand Total)[:\s]*\$?([\d,]+(?:\.\d{2})?)"
# amount_pattern = r"(?:Total|Total Due|Amount Due|Balance Due|Grand Total)[:\s]*\$?([\d,]+(?:\.\d{2})?)"
# # Define regex patterns with proper capturing groups
# priority_amount_pattern = r"(?:Total|Total Due|Amount[\s_]?Due|Total[\s_]?Due|Balance[\s_]?Due|Outstanding[\s_]?Balance|" \
#                           r"Payable[\s_]?Amount|Due[\s_]?Amount|Final[\s_]?Amount|Total[\s_]?Payable|" \
#                           r"Grand[\s_]?Total|Your[\s_]?Responsibility|You[\s_]?Owe|" \
#                           r"Final[\s_]?Charge|Statement[\s_]?Balance|Total[\s_]?Amount[\s_]?Due)[:\s]*\$?([\d,]+(?:\.\d{2})?)"

# date_pattern = r"(?:Due Date|Pay By|Payment Deadline|Bill Date)[:\s]*([\d/.-]+)"
# invoice_pattern = r"(?:Invoice No\.?|Account No\.?|Bill Number)[:\s]*([\w-]+)"




# amount_pattern = r"(?:Total[\s_]?Amount[\s_]?Payable|Amount[\s_]?Due|Payable[\s_]?Amount|Amount[\s_]?Due|Total[\s_]?Due|Balance[\s_]?Due|Outstanding[\s_]?Balance|" \
#                           r"Payable[\s_]?Amount|Due[\s_]?Amount|Final[\s_]?Amount|Total[\s_]?Payable|" \
#                           r"Grand[\s_]?Total|Your[\s_]?Responsibility|You[\s_]?Owe|" \
#                           r"Final[\s_]?Charge|Statement[\s_]?Balance|Total[\s_]?Amount[\s_]?Due)[:\s]*\$?([\d,]+(?:\.\d{2})?)"

# payable_amount_pattern = r"(?:Payable[\s_]?Amount|Final[\s_]?Due|Final[\s_]?Charge)[:\s]*\$?([\d,]+(?:\.\d{2})?)"

amount_pattern = r"(?:Amount[\s_]?Due|Total[\s_]?Due|Balance[\s_]?Due|Outstanding[\s_]?Balance|" \
                 r"Due[\s_]?Amount|Grand[\s_]?Total|Your[\s_]?Responsibility|You[\s_]?Owe|" \
                 r"Statement[\s_]?Balance|Total[\s_]?Amount[\s_]?Due)[:\s]*\$?([\d,]+(?:\.\d{2})?)"

payable_amount_pattern = r"(?:Payable[\s_]?Amount|Final[\s_]?Due|Final[\s_]?Charge|Total[\s_]?Amount[\s_]?Payable)[:\s]*\$?([\d,]+(?:\.\d{2})?)"



fallback_total_pattern = r"(?:Total[\s_]?Amount|Total|Grand[\s_]?Total)[:\s]*\$?([\d,]+(?:\.\d{2})?)"



date_pattern = r"(?:Final[\s_]?payment[\s_]?Deadline|Final[\s_]?payment[\s_]?due[\s_]?by|Payment Due By|Due By|Due Date|Pay By|Payment Deadline|Bill Date|Due[\s_]?Date|DueDate|Payment[\s_]?Due|Pay[\s_]?By|Bill[\s_]?Due|Deadline|Final[\s_]?Payment[\s_]?Date|" \
               r"Last[\s_]?Day[\s_]?to[\s_]?Pay|Total[\s_]?Amount[\s_]?Due[\s_]?By)[\s_]*[:\s]*(\d{4}-\d{2}-\d{2}|\d{1,2}/\d{1,2}/\d{4}|\w+\s\d{1,2},\s\d{4})"


invoice_pattern = r"(?:Account No\.?|Bill Number|Invoice[\s_]?No(?:\.|)|Invoice[\s_]?Number|Inv[\s_]?No|Invoice[#]|Bill[\s_]?No|" \
                  r"Bill[\s_]?ID|Statement[\s_]?No|Reference[\s_]?No|Ref[\s_]?No|Transaction[\s_]?ID|" \
                  r"Order[\s_]?Number|Factura|Invoice[:\s_#]*)[:\s]*([\w\d-]+)"




payment_pattern = r"(Pay[\s_]using|Make[\s_]a[\s_]payment[\s_]at|Make[\s_]payment|Make[\s_]payment[\s_]at|Pay[\s_]at|Pay[\s_]Via|Pay[\s_]Online[\s_]at|Payment[\s_]Link|Call[\s_]to[\s_]Pay|Phone[\s_]Payment|Bank[\s_]Transfer|Payment[\s_]Portal)"

# URL detection for payment links (including .com, .org, etc.)
url_pattern = r"(https?://(?:www\.)?[a-zA-Z0-9-]+\.[a-zA-Z0-9-.]+|www\.[a-zA-Z0-9-]+\.[a-zA-Z0-9-.]+)"

# Function to label text automatically
def auto_label_text(text):
    labeled_tokens = []
    words = text.split()  # Tokenize words

    # Extract entities from text
# Extract entities from text
    payable_amount_match = re.search(payable_amount_pattern, text, re.IGNORECASE)
    amount_match = re.search(amount_pattern, text, re.IGNORECASE)
    date_match = re.search(date_pattern, text, re.IGNORECASE)
    invoice_match = re.search(invoice_pattern, text, re.IGNORECASE)
    payment_match = re.search(payment_pattern, text, re.IGNORECASE)
    url_match = re.search(url_pattern, text)  # Capture any URL


    # Helper function to label multi-word entities
    def label_entity(match, label_type, text):
        if match:
            start, end = match.span()  # Get the start and end positions of the match
            entity_text = text[start:end]  # Extract the matched text
            entity_tokens = entity_text.split()  # Split into tokens
            labels = [f"B-{label_type}"] + [f"I-{label_type}"] * (len(entity_tokens) - 1)  # Assign B- and I- labels
            return list(zip(entity_tokens, labels))  # Return list of (token, label) tuples
        return []

    # Generate labels for each entity type
    labeled_entities = []
    # if amount_match:
    #     labeled_entities.extend(label_entity(amount_match, "Amount", text))
    # if payable_amount_match:
    #   labeled_entities.extend(label_entity(payable_amount_match, "Amount", text))  # Prioritize Payable Amount
    # elif amount_match:
    #   labeled_entities.extend(label_entity(amount_match, "Amount", text))  # If Payable Amount is not found, use Amount Due
    # Extract entities from text

    # Prioritize Payable Amount but also capture Amount Due separately
    if payable_amount_match:
        labeled_entities.extend(label_entity(payable_amount_match, "Amount", text))  # Priority Amount
    elif amount_match and (not payable_amount_match or amount_match.start() != payable_amount_match.start()):
        labeled_entities.extend(label_entity(amount_match, "Amount", text))  # Secondary Amount

    if date_match:
        labeled_entities.extend(label_entity(date_match, "Date", text))
    if invoice_match:
        labeled_entities.extend(label_entity(invoice_match, "InvNum", text))
    if payment_match:
        labeled_entities.extend(label_entity(payment_match, "PayMethod", text))

    # Add URL detection for PayMethod (if a payment-related term exists)
    if url_match:
        labeled_entities.extend(label_entity(url_match, "PayMethod", text))

    # Explicitly handle the "Invoice" keyword, to ensure it's correctly detected even when followed by ID or Number
    explicit_invoice_pattern = r"(Invoice[\s_]*(?:ID|No\.?|Number)[\s_]*[:\s]*)([\w\d-]+)"
    explicit_invoice_match = re.search(explicit_invoice_pattern, text, re.IGNORECASE)
    if explicit_invoice_match:
        labeled_entities.extend(label_entity(explicit_invoice_match, "InvNum", text))  # Explicit invoice handling


    # Create a dictionary for quick lookup
    labeled_entities_dict = {}
    for token, label in labeled_entities:
        cleaned_token = re.sub(r"[:,]", "", token)  # Remove punctuation for matching
        labeled_entities_dict[cleaned_token] = label

    # Assign labels to tokens
    for word in words:
        cleaned_word = re.sub(r"[:,]", "", word)  # Remove punctuation for better matching
        label = labeled_entities_dict.get(cleaned_word, "O")
        labeled_tokens.append((word, label))  # Use original word for printing

    return labeled_tokens

# Example test
sample_text = "Amount Due: $275.50 payment Deadline: 03/15/2025 Bill Number: INV-9087 Pay Online at www.pay.com"
sample_text = "Final Charge: $425.75 Amount Due: $400.00 Payment Due By: 08/20/2024 Invoice No: 12345 Pay using https://www.quickpay.com "
sample_text = "Total Due: $512.30 Payable Amount: $530.00 Due Date: 10/05/2024 Invoice No: 56789 Pay at www.paymentportal.com"
sample_text = "Outstanding Balance: $750.00 Amount Due: $700.00 Payment Deadline: 09/15/2024 Bill Number: BIL-7788 Pay using https://www.paynow.com"
sample_text = "Grand Total: $980.25 Amount Due: $950.00 Pay By: 11/22/2024 Invoice ID: INV-3344 Make a payment at www.quickpay.com"
sample_text = "Final Charge: $630.50 Due Amount: $600.00 Payment Due: 07/01/2025 Account No: ACCT-12345 Pay online at https://www.securepay.com"



labeled_output = auto_label_text(sample_text)

# Print labeled output
for token, label in labeled_output:
    print(f"{token}: {label}")
```
