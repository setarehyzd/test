```python
import json

# Parse the entities string into a dictionary
entities_dict = json.loads(entities)

# Now entities_dict is a dictionary, and we can use .items()
print("Parsed Entities:")
print(entities_dict)
```

```python
import re
import json

# Clean the extracted text
def clean_text(text):
    # Remove extra spaces and special characters
    text = re.sub(r"\s+", " ", text)  # Replace multiple spaces with a single space
    text = re.sub(r"[^\w\s.,-]", "", text)  # Remove special characters except .,-
    return text.strip()

# Map entities to bounding boxes
def map_entities_to_boxes(entities_dict, boxes):
    entity_to_box = {}
    for entity, value in entities_dict.items():
        for box in boxes.split("\n"):
            if value in box:
                entity_to_box[entity] = box
                break
    return entity_to_box

# Parse the entities string into a dictionary
entities_dict = json.loads(entities)

# Clean and preprocess the data
cleaned_text = clean_text(text)
entity_to_box = map_entities_to_boxes(entities_dict, boxes)

# Print the cleaned text and mapped entities
print("Cleaned Text:")
print(cleaned_text)
print("\nParsed Entities:")
print(entities_dict)
print("\nMapped Entities to Bounding Boxes:")
print(entity_to_box)
```

```python
import json

# Save the cleaned text and mapped entities to a JSON file
preprocessed_data = {
    "cleaned_text": cleaned_text,
    "entity_to_box": entity_to_box
}

with open("preprocessed_data.json", "w") as f:
    json.dump(preprocessed_data, f, indent=4)

print("Preprocessed data saved to 'preprocessed_data.json'.")
```
